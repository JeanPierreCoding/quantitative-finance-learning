#### Greek Letter Meanings
* $\omega$: Represents the element of a set. In probability theory it means a speciic outcome from the sample space $\Omega$
* $\epsilon$: "Belongs to"/"is an element of"

## Sets

* **Set**: a collection of distinct objects that are unordered
    set A = {1, 2, 3, 4, a, b, c}
        The information inside of the set is called an element <br>
    * 1 $\epsilon$ A means "1 belongs to A" 
    * 5 $\not\epsilon$ A means "5 does not belong to A"

    * A = {q} (A set with only one element its called a **Singleton**)
    * B = {$\emptyset$} is an **empty set**
    * $\Omega$ = {x, y, z, a, b, c} Omega refers **universal set**
    

## Relations <br>
 * $ \subseteq $ means **subset**
     * $A \subseteq B$ means A is a subset of B <br>
        * A = {1, 2} <br>
        * B = {1, 2 , 3 , 4}
    * If $A \subseteq B$ and $B \subseteq A$, then A = B

    * When we have $\emptyset$ it is by default $\subseteq$ of A

* $\cup$ Union
     * A $U$ B = { $\omega$ $\epsilon$ $\Omega$ : $\omega$ $\epsilon$ A or $\omega$ $\epsilon$ B  } means: <br> 
     "All elements in the universal set belong to either A **or** B"

    * Properties of the Union: <br>
        * A $\cup$ $\emptyset$ = A 
        * A $\cup$ A = A
        * A $\cup$ A $\Omega$ = $\Omega$
        * A $\cup$ B = B $\cup$ A

* $\cap$ Intersection
     * A $\cap$ B {$\omega$ $\epsilon$ $\Omega$ : $\omega$ $\epsilon$ A and $\omega$ $\epsilon$ B}
        * The intersection of sets A and B is the set of all elements that belong to the universal set such that $\omega$ is an element in both A and B
        * More simply: lowercase omega ($\omega$) represents an element that is a part of A and B sets, and snice we're only looking at these that make up the universal set ($\Omega$), it is a part of it. 

     * Properties of the Intersection:
        * A $\cap$ $\emptyset$ = $\emptyset$ 
        * A $\cap$ A = A
        * A $\cap$ A $\Omega$ = A
        * A $\cap$ B = B $\cap$ A

    * A $\cap$ B $\subseteq$ A $\subseteq$ A $\cup$ B means: 
    <br> A intersection of B is essentially a subset of A and it itself is a subset of A union B. 
    <br>
    <br>
        Lets Break it Down!
        * A $\cap$ B $\subseteq$ A: The intersection of A and B is always a subset of A beveause every element in A intersection ($\cap$) B is also in A,
        * A $\subseteq$ A $\cup$ B: The set A is a subset of A union ($\cup$) B, because the unino of the two sets contrain all elements of A (as well as all elements of B)
        * SO, A $\cap$ B is a smaller set within, A which itself is in a larger set of A $\cup$ B. (This establishes a natural ehirarchy of subset relatonships)
            * The intersection of A and B is contained in A, and A is contained in the union of A and B

* Disjoint Sets
    * **Disjoint**: Two or more sets are disjoint if they do not have any elements in common
    * A $\cap$ B = $\emptyset$
        * The intersection of A and B is an empty set, so they share no elements 
    * Example:
        * ( A ) ( B ) <br>
            {1,2}    {4,5}
        * There is no overlap, so they are disjoint

* Partition Sets
    * **Partition**: a partition of a set $S$ is a collection of disjoint subsets that together form the entire set $S$
        * Conditions for a Partition:
            1.  Subsets must be disoint
            2. Union of the subsets must give back the original set 
            3. No subset can be empty
    * Example of a Partition:
    <br>
    Let $S$ = {1, 2, 3, 4, 5, 6}, and we partition into subsets:
        * $A$ = {1, 2}
        * $B$ = {3, 4}
        * $C$ = {5, 6}
        <br>
        So, in union form,
        * $A \cup B \cup C = \{1,2\} \cup \{3,4\} \cup \{5,6\} = \{1,2,3,4,5,6\} = S$

* Complement of Set
    * **Complement**: elements that are not in the given set but exist in the universal set
    * Notation: $\overline{A}$

    * $\overline{A}$ = { $\omega$ $\epsilon$ $\Omega$: $\omega$ $\notin$ A }

    * Properties:
        * $\overline{\overline{A}} = A$
        * $\overline{\emptyset}$ = $\Omega$
        * $A \cup \overline{A} = \Omega$
        * $A \cap \overline{A} = \emptyset$

* Difference of Sets
    * **Difference**: The difference of two sets consists of elements that are in one set but **not** in the other.
    * Notation: $A - B$ or $A \setminus B$
    * Definition: $A - B = \{ x \in A \mid x \notin B \}$
        * This means that $A - B$ consists of all elements that are in $A$ but **not in** $B$.

    * Example of a Set Difference:
    <br>
    Let:
        * $A = \{1, 2, 3, 4, 5\}$
        * $B = \{3, 4, 6\}$
    <br>
    Then:
        * $A - B = \{1,2,5\}$ (Elements in $A$ but **not** in $B$)
        * $B - A = \{6\}$ (Elements in $B$ but **not** in $A$)

    - **Key Properties of Set Difference**
        - **Non-Commutative**:  
        $$
        A - B \neq B - A
        $$
        The difference depends on the order of the sets.

        - **Difference with Itself**:  
        $$
        A - A = \emptyset
        $$
        The difference of a set with itself is always the empty set.

        - **Difference with the Universal Set**:  
        $$
        A - \Omega = \emptyset
        $$
        If you subtract the universal set, nothing remains.

        - **Set Difference and Intersection Relationship**:  
        $$
        A - B = A \cap B^c
        $$
        - The difference of two sets is the intersection of \( A \) with the **complement** of \( B \).
