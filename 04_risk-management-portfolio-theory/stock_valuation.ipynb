# Stock Valuation

Stocks are securities that represent ownership and are typiclaly called **Common Stock**:
- Give their owners certain rights
- limit the owner's liability in the event of bankruptcy,
- Owners of common stock of a company are called shareholders
- Number of shares held by an individual in relation to the total number issued by a company represents the individuals proportional Ownership of the company. 

## Key Concepts:
- **Common Stock Par Value** : Face value of the stock, and usually has no relation to the market value. Some companies have shares with 0 par value, so a stock in the market can sell for $100 a share and have par value of 0 in the copmany books
- **Dividend**: Payment by a company to the shareholdes that represents a return on cpital provided by the shareholder to the company.
- **Market Value**: Each share of stock is entitled to one vote to elect director at annual meeting of the company.
    - "In straight voting, one vote per share is cast for each director up for election.
    - In cumulative voting, all the votes available to each shareholder (depending on the number of directors up for election) can be cast for only one director (rather than one for each director).
    - For example, if you own 100 shares of a company and the company is electing 3 directors, you have 300 votes. In straight voting you can cash 100 votes for each director. In cumulative voting you can cash all 300 votes for one director." (Yash R. Puri, Stock Valuation)
- **Preemptive Right**: If a company issues new shares, the existing shareholders have a right to purchase the new shares in proprotion to what they hold.
    - "For example, if you own 5% of a companyâ€™s shares, and the company decides to issue 1,000 new shares to raise new capital, preemptive right will give you the right to purchase 50 new shares issued by the company." (Yash R. Puri, Stock Valuation)

## Valuation Principle
- Value of a security is based on the cash flows provided by the security to the purchaser

### Bonds v. Stocks
- Bond valuation relatively easy as cash flows are known at the time of purchase
    - Interest payment during the life of the bond, and
    - repayment of the face value at maturity

- Stock valuation is harder since cash flows from stocks to the investor are not certain
    - Dividends are not guarantee; nor do they remain fixed
    - The investment is forever, and the company is not required to pay back the investment at any time

### Simple Stock Valuation Model
- Simple assumptions to estimate future cash flows, and
    - calculate their present value to arrive at stock value

- We will use the following terminology:
    - **$P_t$**: **Market Price** per share at the end of time period $t$
    - **$D_t$**: **Dividend** per share at the end of pr\period $t$, where
    - **$t$**: the time period (0, 1, 2, etc.)
    - **$R$**: **Market required rate of return** for investment in this risk class

#### Simple Cash Flow Assumptions
- You plan to buy a share of stock today, and you know what the stock:
    - will pay a dividend of $5 a year from now, and
    - will be worth $85 at that time
- Assume you require 10% rate of return from such stock, how much is the stock worth to you today?
    - Note: Real life, we don't know these values. These assumptions are for arriving to a methodology for valuing stocks

##### Stock Cash Flows
- If you purchase the stock, you receive:
    - $P_0$ = ?
    - $R$ = 20%
    - $D_1$ = $5
    - $P_1$ = $85
- We know that at the end of the year 1, the stock will be worth $85 and that we will receive a dividend of $5
- We know that our required rate of return is %20
- We will need to calculate Present Value with the variables we have
$$ PV_0 = \frac{FV_n}{(1+R)^n} $$

So, the PV of the cash flows is:
$$ PV_0 = \frac{D_1+P_1}{(1+R)^1} = \frac{5+85}{(1+.2)^1} = \frac{90}{(1.2)^1} = 75$$
Therefore, based on the expected cashflow and your required rate of return, the stock is worth $75

<br>
<br>
<br>

- Now, imagine that you have traveled forward one year in time! At that time you find that the same stock 
    - Will pay another $5 per share at the end of the year from then, and
    - will have a market price of $97 at that time. 
    - How much is the stock worth given your required rate of return of 20% when you arrive at one-year-forward in time?
    <br>
    <br>
    - $P_1$ = ?
    - $R$ = 20%
    - $D_2$ = $5
    - $P_2$ = $97
    <br>
So, the PV of the cash flows is:
$$ PV_1 = \frac{D_2+P_2}{(1+R)^1} = \frac{5+97}{(1+.2)^1} = \frac{102}{(1.2)^1} = 85$$
Therefore, based on the expected cashflow and your required rate of return, the stock is worth $85
<br>
- Substituting the formula for $P_1$, we get:
$$
P_0 = \frac{D_1 + \frac{D_2 + P_2}{(1+R)}}{(1+R)}
$$

$$
P_0 = \frac{D_1}{(1+R)^1} + \frac{D_2}{(1+R)^2} + \frac{P_2}{(1+R)^2}
$$


- The value of the stock is the present value of dividends plus the value at the end of the time horizon.

- If we again go forward another year, we use the info about price and dividend at end of year three to find $P_2$
$$
P_2 = \frac{D_3 + P_3}{(1+R)}
$$

$$
P_0 = \frac{D_1}{(1+R)^1} + \frac{D_2}{(1+R)^2} + \frac{P_2}{(1+R)^2} + \frac{P_2}{(1+R)^2}
$$


- **Thus, the stock price is the PV of all future dividends**
    - Its hard however to forecast dividends, so simplifying assumptions are used

### No Growth Model

* When there is no growth, all dividends are the same:
    - $ D_1 = D_2 = D_3 ... = D $ (a constant)

* If R is required rate of return, the stock price formula simplifies to:
    - $ P_0 = \frac{D}{R} $
