### Risk-Return Arbitrage
* Suppose A and B provide the same returm what will investors do?
  - Arbitrage by simultaneously selling B and buying A (same return but with lower risk)
  - Since investors are selling B, the price will fall; and since they are buying A, the price will increase
  - As more people buy A, the price will go up, and a higher price means lower return, and as more people sell B, the price falls which means a higher return
  - So, lower risk project A will provide lower return and higher risk project B will provide higher return
  - Equilibrium prices setttles when the investors believe that the rates of return from A and B are commensurate with their riskiness. In the end the safter project offers a lower return and riskier project offers a higher return
* Risk-Return arbitrage ensures higher risk projects provide higher expected returns

### Project or Stand-Alone Risk
* Risk: variability of future possible returns
* Stand-alon or project risk: risk from investing in a single asset
* Statistical techniques that measure variability are used to estimate risk
  - We combine two measures:
    - Expected value, and
    - Standard deviation
  - We will use discrete variables for simplicity

### Measuring Project Risk
* Cash outflow in the furure depend onf the state of the economy
* Lets define the following:
  - $ X_i $ as a random variable where $i$ represents the state of the economy (e.g. recession or booming), and
  - $ P_i $ as the probability of state i occuring
* Expected value of the outcome is
$$
% Expected Value
E(X) = \sum_{i=1}^{n} P_i \, X_i
$$
  - where $n$ is the number of states possible
* The variance of cashflows is
$$
% Variance
\mathrm{Var}(X) = \sum_{i=1}^{n} P_i \,\bigl(X_i - E(X)\bigr)^2
$$

* Standard deviation is the square-root of the variance
$$
\sigma_X = \sqrt{\mathrm{Var}(X)}
          = \sqrt{\sum_{i=1}^{n} P_i \bigl(X_i - E(X)\bigr)^2}.
$$
  - Measures the dispersion of cash flows; high dispersion means that possible cash flows are widely distributed meaning higher risk

#### Example: Where only four outcomes are possible, we get:
  - $ \E(X) = P_1 X_1 + P_2 X_2 + P_3 X_3 + P_4 X_4 $

and 

- $ \sigma_X = \sqrt{ P_1 \bigl(X_1 - \E(X)\bigr)^2 + P_2 \bigl(X_2 - \E(X)\bigr)^2 + P_3 \bigl(X_3 - \E(X)\bigr)^2 + P_4 \bigl(X_4 - \E(X)\bigr)^2}. $
