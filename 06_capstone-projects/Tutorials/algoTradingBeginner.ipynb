```python
close=stocks.loc[:,"Close"].copy()
close
```

```python
%matplotlib inline
plt.style.use("seaborn")
```

```python
close.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
close.iloc[0,0]
```

```python
close.AAPL
```

```python
# Normalize data
# Divides all value with starting point value by using .div() and the multiply by 100 .mul()
close.AAPL.div(close.iloc[0,0]).mul(100)
```

```python
# Normalize data for all stocks at once
# Divides all value with starting point value by using .div() and the multiply by 100 .mul()
close.div(close.iloc[0]).mul(100)
```

```python
normclose=close.div(close.iloc[0]).mul(100)
```

```python
normclose.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
